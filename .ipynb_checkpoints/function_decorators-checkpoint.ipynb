```python
def iteration_average(seq:list)->float:
    """ Returns the mean of a sequence 
        >>> iteration_average([1,2,5])
        2.6666666666666665
        """
    return sum(seq)/len(seq)

iteration_average([23,2,5,-8])
```

```python
def iteration_average_f(seq:list)->float:
    """ Returns the mean of a sequence 
        >>> iteration_average_f([1,2,5])
        2.66666666666665
        """
    return sum(seq)/len(seq)

iteration_average_f([23,2,5,-8])
```

```python
def range_inner(data,groupsize=4,group=[]):
    ranges = []
    it = iter(data)
    while True:
        try:
            group.append(next(it))
        except StopIteration:
            break
        if len(group) == groupsize:
            ranges.append(iteration_average_f(group))
            group = []
    return ranges
```

```python
def range_inner_apply(data,groupsize=4,group=None):
    group = group if group is not None else []
    ranges = []
    it = iter(data)
    while True:
        try:
            group.append(next(it))
        except StopIteration:
            break
        if len(group) == groupsize:
            ranges.append(iteration_average_f(group))
            group = []
    return ranges
```

```python
def funRange(data,groupsize=4,group=None,aggregates=iteration_average_f):
    group = group if group is not None else []
    ranges = []
    it = iter(data)
    while True:
        try:
            group.append(next(it))
        except StopIteration:
            break
        if len(group) == groupsize:
            ranges.append(aggregates(group))
            group = []
    return ranges
```

```python
range_inner_apply([1,2,3,4,5,6])
```

```python
funRange([1,2,3,4,5,6])
```

```python
from typing import Callable
def closure_ex(aggregates:Callable)->Callable:
    def wrapper_ex(data,*,groupsize=4,group=None):
        group = group if group is not None else []
        ranges = []
        it = iter(data)
        while True:
            try:
                group.append(next(it))
            except StopIteration:
                break
            if len(group) == groupsize:
                ranges.append(aggregates(group))
                group = []
        return ranges
    return wrapper_ex
closure_ex(iteration_average)
closure_ex(range_inner_apply)
```

```python
closure_fun_one = closure_ex(iteration_average)
closure_fun_two = closure_ex(range_inner_apply)
```

```python
closure_fun_one([0,1,2,3,4,5])
```

```python
closure_fun_two(range(21))
```

```python
type(closure_fun_two)
```

```python
type(closure_fun_one)
```

```python
def myArg(*argv):
    for arg in argv:
        print(arg)
  
  
myArg('Hello', 'Welcome', 'to', 'Multi', 'argument', 'input')
```

```python
def myKwargs(arg1, **kwargs):
    for key, value in kwargs.items():
        print("%s == %s" % (key, value))
  
  
# Driver code
myKwargs("Hi", first='Multi', mid='args', last='input')
```

```python
def myArgsKwargs(*argv, **kwargs):
    print("-----for Args-----")
    for arg in argv:
        print(arg)
    print("-----for KwArgs-----")    
    for key, value in kwargs.items():
        print("%s == %s" % (key, value))
  
  
# Driver code
myArgsKwargs("Hi", 'Vinod','How','Are','You?',first='Multi', mid='args', last='input')
```

```python
def seperateArgsKwargs(func:Callable)->Callable:
    def wrapper(*args,**kwargs):
        resp = func(*args,**kwargs)
        return resp
    return wrapper

Args = seperateArgsKwargs(myArg('Hello', 'Welcome', 'to', 'Multi', 'argument', 'input'))
print(Args)
kwarg = seperateArgsKwargs(myKwargs('Hi',first='Multi', mid='args', last='input'))
print(kwarg)
```
